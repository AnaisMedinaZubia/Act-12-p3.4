## Las variables
1.  CT = --> Costo de traje
2.  DE = --> Descuento que se obtendrá
3.  PF = --> Precio final del traje

## Formulas
* $DE = CT * 0.15$
* $DE = CT * 0.08$

## Diagrama de flujo lucidchart
<img src="https://raw.githubusercontent.com/AnaisMedinaZubia/Act-12-p3.4/refs/heads/main/DF%20costodetraje.png" width="400" height="500" alt="Texto descriptivo">


```python
# Para cuando sea verdadero(Si)
# Calcular cuanto costara pagar un traje if else
# Datos de entrada CT, DE, PF
# Se compara si el costo del traje es mayor a $2500.
# Se calcula el descuentoaplicando 8% e ser falso
# Se calcula el descuento aplicando 15 % de ser verdadero
CT = int(input("Ingrese el costo del traje: "))
if CT > 2500:
    DE = CT * 0.15
else:
    DE = CT * 0.08
PF = CT - DE
print("El costo del traje es: ", CT)
print("El descuento es: ", DE)
print("El precio final es: ", PF)

print("Fin del programa")
print("Programa realizado por Anais Medina NL = 36")
```
